## Challenge 1 - Most Profiting Authors

### Step 1: Calculate the royalties of each sales for each author

```python
%%sql
select *
from sales
where title_id='MC3021'
```

```python
%%sql
select t.title_id Title_ID, a.au_id AUTHOR_ID, (t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) Royalty
from 
    authors a
    left join 
        (titleauthor ta   
        inner join 
        (titles t left join sales s on t.title_id = s.title_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id

order by Title_ID,AUTHOR_ID
;
```

### Step 2: Aggregate the total royalties for each title for each author

```python
%%sql
with step1 as (
select t.title_id Title_ID, a.au_id AUTHOR_ID, (t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) Royalty
from 
    authors a
    left join 
        (titleauthor ta   
        inner join 
        (titles t left join sales s on t.title_id = s.title_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id
)
select Title_ID, AUTHOR_ID, sum(Royalty) Total_Royalty
from step1
group by AUTHOR_ID, Title_ID
;
```

### Step 3: Calculate the total profits of each author

```python
%%sql
with step2 as (
with step1 as (
select t.title_id Title_ID, a.au_id AUTHOR_ID, (t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) Royalty
from 
    authors a
    left join 
        (titleauthor ta   
        inner join 
        (titles t left join sales s on t.title_id = s.title_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id
)
select Title_ID, AUTHOR_ID, sum(Royalty) Total_Royalty
from step1
group by AUTHOR_ID, Title_ID
)
select AUTHOR_ID, (Total_Royalty+t.advance*ta.royaltyper/100) PROFITS
from step2 s2
    inner join 
    (titles t left join titleauthor ta on t.title_id = ta.title_id)
    on s2.Title_ID = t.title_id and s2.AUTHOR_ID=ta.au_id
order by PROFITS DESC
limit 3
;

```

## Challenge 2

```python
%%sql 

CREATE TABLE most_profiting_authors as
with step2 as (
with step1 as (
select t.title_id Title_ID, a.au_id AUTHOR_ID, (t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) Royalty
from 
    authors a
    left join 
        (titleauthor ta   
        inner join 
        (titles t left join sales s on t.title_id = s.title_id)
        on ta.title_id = t.title_id)
    on a.au_id = ta.au_id
)
select Title_ID, AUTHOR_ID, sum(Royalty) Total_Royalty
from step1
group by AUTHOR_ID, Title_ID
)
select AUTHOR_ID au_id, (Total_Royalty+t.advance*ta.royaltyper/100) PROFITS
from step2 s2
    inner join 
    (titles t left join titleauthor ta on t.title_id = ta.title_id)
    on s2.Title_ID = t.title_id and s2.AUTHOR_ID=ta.au_id
;
```

```python
%%sql
select *
from most_profiting_authors
order by PROFITS desc
```
